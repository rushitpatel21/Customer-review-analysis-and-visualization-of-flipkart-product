```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import re

def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)

excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = 'Flipkart customer review'
sheet.append(['No.','Title','Desc', 'Star', 'Date'])

allData = []
```

```python
def getData(URL):
    r = requests.get(URL)
    soup = BeautifulSoup(r.text,'html.parser')
    return soup

def getNextPage(soup):
    checkNextPage = soup.find('a',class_='_1LKTO3')
    if checkNextPage != None:
        allComments = soup.find_all('div',class_='_1AtVbE')
        for comment in allComments:
            title = comment.find('p',class_='_2-N8zT')
            desc = comment.find('div',class_='t-ZTKy')
            star = comment.find('div',class_='_3LWZlK _1BLPMq')
            date = comment.find('p',class_='_2mcZGG')
            if(title != None and desc != None and star != None and date != None):
                print(remove_emojis(title.text), remove_emojis(desc.text), int(star.text), date.next_sibling.next_sibling.text)
                allData.append([title.text, desc.text, int(star.text), date.next_sibling.next_sibling.text])
    else:
        return 'stop'
```

```python
num = 1
while True:
    
    soup = getData('https://www.flipkart.com/hp-g200-wired-optical-gaming-mouse/product-reviews/itm365abb464dc38?pid=ACCFZPG3N6QEXUUM&lid=LSTACCFZPG3N6QEXUUMU6JXHR&marketplace=FLIPKART&page='+str(num))
    checkStop = getNextPage(soup)
    num = num + 1 
    if checkStop == 'stop':
        break
```

```python
for index, data in enumerate(allData):
    sheet.append([index+1, remove_emojis(data[0]), remove_emojis(data[1]), data[2], data[3]])
```

```python
excel.save('FlipkartReview.csv')
```
