```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('./FlipkartReview.csv')
```

```python
def handleDate(data):
    if 'ago' in data:
        return np.nan
    else:
        return data
    
df['Date'] = df['Date'].apply(handleDate)
```

```python
def handleDesc(data):
    if 'READ MORE' in data:
        return data.replace('READ MORE','')
    else:
        return data
    
df['Desc'] = df['Desc'].apply(handleDesc)
```

```python
graph1 = pd.DataFrame({'Star':df['Star'],'Date':df['Date']})
graph1
```

```python
sns.lineplot(x='Star',y='Date',data=graph1)
```
